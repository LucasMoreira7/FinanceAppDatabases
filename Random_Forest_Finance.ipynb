```python
#Conectando o notebook com o Drive
from google.colab import drive
drive.mount('/content/gdrive')
from google.colab import auth
import gspread

## Importando bibliotecas necessárias
import pandas as pd
from datetime import datetime, timedelta
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.impute import SimpleImputer
import numpy as np

# Carregar o DataFrame
data_combined = pd.read_csv('/content/gdrive/MyDrive/FinanceDatabase/BaseCompletaTratada.csv')
display(data_combined)
```

```python
# Especificar as features que serão usadas para a previsão
features = [col for col in data_combined.columns if col not in ['preco', 'dia_da_semana', 'data']]
#features = [col for col in data_encoded.columns if col not in ['data', 'categoria', 'despesaOuReceita']]

# Definir a variável target que queremos prever
target = 'preco'

# Dividir os dados em conjuntos de treinamento e teste
train_data, test_data = train_test_split(data_combined, train_size=0.7, random_state=42)

# Criar um objeto imputer para preencher os valores ausentes com a média
imputer = SimpleImputer(strategy='constant', fill_value=1)

# Aplicar o imputer nos dados de treinamento
train_data[features] = imputer.fit_transform(train_data[features])

# Aplicar o imputer nos dados de teste
test_data[features] = imputer.transform(test_data[features])

# Criar o modelo de regressão com Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Treinar o modelo com os dados de treinamento
rf_model.fit(train_data[features], train_data[target])

# Fazer previsões nos dados de teste
predictions = rf_model.predict(test_data[features])

# Avaliar o desempenho do modelo com algumas métricas
mae = mean_absolute_error(test_data[target], predictions)
mse = mean_squared_error(test_data[target], predictions)
r2 = r2_score(test_data[target], predictions)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("R-squared (R²):", r2)
```
